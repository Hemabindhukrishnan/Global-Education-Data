```python
# bar chart(1)

val1=globaled1["Youth_15_24_Literacy_Rate_Male"]
val2=globaled1["Youth_15_24_Literacy_Rate_Female"]
cate=globaled1["Countries and areas"]

#plotting
fig,axes=plt.subplots(figsize=(17,12),dpi=144)

bar_width=0.35
index=range(len(cate))
bar1=axes.bar(index,val1,bar_width,label="Youth_15-24_Age_Literacy_Rate_Male",color='black')
bar2=axes.bar([i+bar_width for i in index],val2,bar_width,label="Youth_15-24 Age_Literacy_Rate_Female")

#adding labels,tittle,and custom x-axis tick labels

axes.set_xlabel('countries',fontsize=29.0,fontweight="bold")
axes.set_ylabel('values',fontsize=29.0,fontweight="bold")
axes.set_title('Comparison between males and Female (Youth_15-24_Age_Literacy_Rate)',fontsize=29.0,color="red",fontweight="bold")
axes.set_xticks([i+bar_width/2 for i in index])
plt.xticks(rotation=90,fontsize=24.0)
plt.yticks(fontsize=25.0)

axes.set_xticklabels(cate,fontweight="bold",fontsize=24)
plt.grid(True)

#Addding legend
axes.legend(loc='upper right',fontsize=25)

#show plot
plt.show()

#-----------------------------------------------------

#barchart(2)

val3=globaled1["OOSR_Upper_Secondary_Age_Male"]
val4=globaled1["OOSR_Upper_Secondary_Age_Female"]
cate=globaled1["Countries and areas"]
#plotting
fig,axes=plt.subplots(figsize=(17,12),dpi=144)

bar_width=0.35
index=range(len(cate))
bar3=axes.bar(index,val3,bar_width,label="OOSR_Upper_Secondary_Age_Male",color='Red')
bar4=axes.bar([i+bar_width for i in index],val4,bar_width,label="OOSR_Upper_Secondary_Age_Female",color='Green')

#adding labels,tittle,and custom x-axis tick labels

axes.set_xlabel('countries',fontsize=29,fontweight="bold")
axes.set_ylabel('values',fontsize=29,fontweight="bold")
axes.set_title('Comparison between males and Female (OOSR_Upper_Secondary_Age)',fontsize=29.0,color="red",fontweight="bold")
axes.set_xticks([i+bar_width/2 for i in index])
plt.xticks(rotation=90,fontsize=24.0)
plt.yticks(fontsize=25.0)

axes.set_xticklabels(cate,fontweight="bold")
plt.grid(True)
#Addding legend
axes.legend(loc='upper right',fontsize=25)

#show plot
plt.show()

#-----------------------------------------------------
#barchart(3)

val5=globaled1["Completion_Rate_Upper_Secondary_Male"]
val6=globaled1["Completion_Rate_Upper_Secondary_Female"]
cate=globaled1["Countries and areas"]

#plotting
fig,axes=plt.subplots(figsize=(17,12),dpi=144)

bar_width=0.35
index=range(len(cate))
bar5=axes.bar(index,val5,bar_width,label="Completion_Rate_Upper_Secondary_Male",color='Yellow')
bar6=axes.bar([i+bar_width for i in index],val6,bar_width,label="Completion_Rate_Upper_Secondary_Female",color='Orange')

#adding labels,tittle,and custom x-axis tick labels

axes.set_xlabel('countries',fontsize=29,fontweight="bold")
axes.set_ylabel('values',fontsize=29,fontweight="bold")
axes.set_title('Comparison between males and Female (Completion_Rate_Upper_Secondary)',fontsize=29.0,color="red",fontweight="bold")
axes.set_xticks([i+bar_width/2 for i in index])
plt.xticks(rotation=90,fontsize=24.0)
plt.yticks(fontsize=25.0)
axes.set_xticklabels(cate,fontweight="bold")
plt.grid(True)
#Addding legend
axes.legend(loc='upper right',fontsize=25)

#show plot
plt.show()

```

```python

fig,axes = plt.subplots(figsize=(29,30))
globaled2=globaled.head(15)

countries=globaled2["Countries and areas"]
values1=globaled2["Completion_Rate_Upper_Secondary_Male"]
values2=globaled2["Completion_Rate_Upper_Secondary_Female"]
values3=globaled2["OOSR_Upper_Secondary_Age_Male"]
values4=globaled2["OOSR_Upper_Secondary_Age_Female"]
values5=globaled2["Youth_15_24_Literacy_Rate_Male"]
values6=globaled2["Youth_15_24_Literacy_Rate_Female"]

axes.scatter(countries,values1,label="completion_rate_Upper_secondary_male",color="purple",s=290)
axes.scatter(countries,values2,label="completion_rate_upper_secondary_female",color="blue",s=290)
axes.scatter(countries,values3,label="OOSR_Upper_Secondary_Age_Male",color="red",marker='s',s=290)
axes.scatter(countries,values4,label="OOSR_Upper_Secondary_Age_Female",color="Green",marker='s',s=290)
axes.scatter(countries,values5,label="Youth_15_24_Literacy_Rate_Male",color="Black",marker='^',s=290)
axes.scatter(countries,values6,label="Youth_15_24_Literacy_Rate_Female",color="orange",marker='^',s=290)

axes.set_xlabel('Countries',color='red',fontsize=55)
axes.set_ylabel('Frequency',color='red',fontsize=55)
plt.xticks(rotation=90,fontsize=40.0,fontweight="bold")
plt.yticks(fontsize=40.0,fontweight="bold")
plt.legend(fontsize=35.5,loc="right")
plt.title('Scatter plot of Counties Educational details',color="Red",fontsize=59.0)
plt.grid(True,color="grey")
plt.show()
```

```python
#heat map
comparison=['Completion_Rate_Upper_Secondary_Male','Completion_Rate_Upper_Secondary_Female','OOSR_Upper_Secondary_Age_Male','OOSR_Upper_Secondary_Age_Female','Youth_15_24_Literacy_Rate_Male','Youth_15_24_Literacy_Rate_Female']
corelation=globaled2[comparison].corr()
plt.figure(figsize=(50,50))
sns.heatmap(corelation,annot=True,cmap='coolwarm',fmt=".2f",linewidths=2.8,annot_kws={"weight":"bold"})
sns.set(font_scale=8.5)
plt.xticks(fontweight="bold",fontsize=85.5)
plt.yticks(fontweight="bold",fontsize=85.5)
plt.title('Heatmap for Global education data',fontsize=120.0,fontweight="bold",color="red")
plt.show()
```
